```python
import os
import pandas as pd
```

```python
PATH = "case_studies"
```

```python
data = []

for filename in os.listdir(PATH):
    if filename.endswith(".txt"):
        file_path = os.path.join(PATH, filename)
        
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()
        
        data.append({
            "title": filename,
            "text": text
        })

df = pd.DataFrame(data)

print(df.head())
```

```python
df.to_parquet("case_studies.parquet", index=False)
```
